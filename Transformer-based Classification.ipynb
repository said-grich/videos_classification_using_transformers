## Data Collection

#### Getting videos names

```python
video_path = "./Videos"
video_names = os.listdir(video_path)
print(video_names)

print('Number of videos =', len(video_names))
```

#### Preparing dataset

```python
df = pd.read_excel('DataSource.xlsx') 
dataset = df[['points','Cu (%)','Fe (%)','Pb (%)','Zn (%)']]

# selecting names where name is in videos names
result = dataset.loc[dataset['points'].isin(video_names)]
  
print('\nResult dataframe :\n', result)
print('\nNumber of rows :\n', len(result))
```

```python
# videos of Cleaner 1 (rl1)
rl1 = result[result['points'].str.contains("rl1")]

print('\nResult dataframe :\n', rl1)
print('\nNumber of rows :\n', len(rl1))
```

```python
# videos of Cleaner 2 (rl2)
rl2 = result[result['points'].str.contains("rl2")]

print('\nResult dataframe :\n', rl2)
print('\nNumber of rows :\n', len(rl2))
```

```python
# videos of Cleaner 3 (rl3)
rl3 = result[result['points'].str.contains("rl3")]

print('\nResult dataframe :\n', rl3)
print('\nNumber of rows :\n', len(rl3))
```

```python
# videos of Cleaner 4 (rl4)
rl4 = result[result['points'].str.contains("rl4")]


print('\nResult dataframe :\n', rl4)
print('\nNumber of rows :\n', len(rl4))
```

#### Tagging

```python
def label_race (row):
    
    
    
    if row['points'].startswith("rl1"):
        
        if row['Pb (%)'] < 17 :
            return 'Very Low'
        if row['Pb (%)'] >= 17 and row['Pb (%)'] < 20 :
            return 'Low'
    
        if row['Pb (%)'] >= 20 and row['Pb (%)'] < 22 :
            return 'Medium'
        
        if row['Pb (%)'] >= 22 and row['Pb (%)'] < 24 :
            return 'High'

        if row['Pb (%)'] >= 24 :
            return 'Very High'
        
        
    if row['points'].startswith("rl3"):
        
        if row['Pb (%)'] < 36 :
            return 'Very Low'
        
        if row['Pb (%)'] >= 36 and row['Pb (%)'] < 39 :
            return 'Low'
    
        if row['Pb (%)'] >= 39 and row['Pb (%)'] < 42 :
            return 'Medium'
        
        if row['Pb (%)'] >= 42 and row['Pb (%)'] < 45 :
            return 'High'

        if row['Pb (%)'] >= 45 :
            return 'Very High'
        
        
        
        
    if row['points'].startswith("rl4"):
        
        if row['Pb (%)'] < 45 :
            return 'Very Low'
        
        if row['Pb (%)'] >= 45 and row['Pb (%)'] < 50 :
            return 'Low'
    
        if row['Pb (%)'] >= 50 and row['Pb (%)'] < 55 :
            return 'Medium'
        
        if row['Pb (%)'] >= 55 and row['Pb (%)'] < 60 :
            return 'High'

        if row['Pb (%)'] >= 60 :
            return 'Very High'
        
        
    else:
        return 'Unkonwn'
        
        
    
   
 
```

```python
result_tagged = result
result_tagged['Tag'] = result_tagged.apply (lambda row: label_race(row), axis=1)
result_tagged.rename(columns = {'points':'Video_name'}, inplace = True)
result_tagged
```
